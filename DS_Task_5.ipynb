```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load Dataset
df = pd.read_csv("accidents.csv")
print(df.head())
print("Shape:", df.shape)

# 2. Extract Hour from time
df['Hour'] = pd.to_datetime(df['Start_Time']).dt.hour

# 3. Plot 1 - Accidents by Hour (Time of Day)
plt.figure(figsize=(8, 4))
df['Hour'].value_counts().sort_index().plot(kind='bar', color='steelblue')
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 4. Plot 2 - Accidents by Weather Condition
plt.figure(figsize=(7, 4))
df['Weather_Condition'].value_counts().plot(kind='bar', color='orange')
plt.title("Accidents by Weather Condition")
plt.xlabel("Weather")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 5. Plot 3 - Accidents by Road Condition
plt.figure(figsize=(6, 4))
df['Road_Condition'].value_counts().plot(kind='bar', color='red')
plt.title("Accidents by Road Condition")
plt.xlabel("Road Condition")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 6. Plot 4 - Accidents by Severity
plt.figure(figsize=(6, 4))
df['Severity'].value_counts().sort_index().plot(kind='bar', color='purple')
plt.title("Accidents by Severity Level")
plt.xlabel("Severity (1=Low, 4=High)")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
